# How to optimize Python while functional programming

Ther are many ways to implement a dotproduct of two vectors in Python. dotproduct for `v1 = (3, 4)` and `v2 = (5, 1)` would be `dp = 3*5 + 4*1`.

The code below implements it fairly cleanly in Python:

```python
import operator

def dotproduct(vec1, vec2, mul=operator.mul):
    return sum(map(mul, vec1, vec2))
```

However it can be optimised significantly if written like so:

```python
def dotproduct(vec1, vec2, sum=sum, map=map, mul=operator.mul):
    return sum(map(mul, vec1, vec2))
```
