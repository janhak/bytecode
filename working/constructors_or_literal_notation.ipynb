# What is faster constructor or literal notation?

Most data structure in Python can be initialized using either literal notation like: 
- `{1, 2}` to define a new set containing two elements
or a constructor:
- `set((1, 2))`
both of these are equvalent in terms of results, but one is significantly faster, run the snippet below and check which:

```python
from timeit import timeit
print('Literal notation: ', timeit('{1, 2}'))
print('Constructor: ', timeit('set((1, 2))'))
```
