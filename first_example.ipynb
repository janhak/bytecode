# What the compiler doesn't know

The compilation stage in contrast to the interpretation stage, is a fairly lightweight process with Python. The compiler is really suprisingly simple.

```python
def mod(a, b):
    return a % b
```

```python
dis.dis(mod)
```

```python
mod(12, 5)
```

```python
mod('by%sde', 'teco')
```

The examples above illustrate the difficulty of optimizing the bytecode at compilation stage. We have no information about the nature of the object that we will be operating on as Python is a dynamically typed language. Even simple modulo division means that internaly python calls the `__mod__` method of the object a: `a.__mod__(b)`. When the object is `str` the modulo division operator for it does something entirely differnt a string interpolation, the more familiar syntax of: `'by%sde' & 'teco'`.

```python
def mod(a, b):
    a % b
    return a % b
```

```python
dis.dis(mod)
```

This illustrates further troubles even though first `a % b` does nothing there is really no way of knowing it at compilation time. As we already witnessed above this will just call a `__mod__` method on an object and this method might communicate with the database, write to file or do all other sorts of things. This means that both of this calls will be executed at runtime.
