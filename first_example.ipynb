```python
def f1(x):
    if x < 0:
        return 'Negative'
    elif x == 0:
        return 'Zero'
    elif x > 0:
        return 'Positive'
    
def f2(x):
    if x < 0:
        return 'Negative'
    else:
        if x == 0:
            return 'Zero'
        else:
            if x > 0:
                return 'Positive'
```

```python
f1.__code__.co_code == f2.__code__.co_code
```
