## How different code can be the same ByteCode

Consider the functions below:

```python
def f1(x):
    if x < 0:
        return 'Negative'
    elif x == 0:
        return 'Zero'
    elif x > 0:
        return 'Positive'
    
def f2(x):
    if x < 0:
        return 'Negative'
    else:
        if x == 0:
            return 'Zero'
        else:
            if x > 0:
                return 'Positive'
```

They are written differently but execute same logic. If you the `co_code` attribute is the bytecode for the function object. Execute the line below, what does it tell you about their bytecode? Get the dis module out and compare...

```python
f1.__code__.co_code == f2.__code__.co_code
```
